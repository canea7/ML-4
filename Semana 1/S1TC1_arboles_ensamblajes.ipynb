```python
# Atípicos

# Número de desviaciones atípicas
num_desv_tip = 4

# Función para cálcular los índices de los atípicos
def atipicos_desv_tip(variable, num_desv_tip = 3):
    #sacamos los nulos por ahora
    variable = variable.dropna()
    
    #calculamos los límites
    media = np.mean(variable)
    sd = np.std(variable)
    umbral = sd * num_desv_tip
    lim_inf = media - umbral
    lim_sup = media + umbral
    
    #encontramos los índices de los que están fuera de los límites
    indices = [indice for indice,valor in variable.items() if valor < lim_inf or valor > lim_sup]
    return(indices)

# Función para contar el número de atípicos
def conteo_atipicos(df,variable, num_desv_tip = 3):
    atipicos = atipicos_desv_tip(df[variable], num_desv_tip)
    return(df.loc[atipicos,variable].value_counts())

# Lista de variables
columns_names = bikes.columns.values
columns_names = list(columns_names)

# Aplicar función de conteo atípicos
for variable in columns_names:
    print('\n' + variable)
    print(conteo_atipicos(bikes,variable,num_desv_tip))
    #print('')
```

### Punto 4 - Árbol de decisión manual
En la celda 4 cree un árbol de decisiones para pronosticar la variable "total" iterando **manualmente** sobre las variables "hour" y  "season". El árbol debe tener al menos 6 nodos finales.

```python
# Celda 4
# Definición de parámetros y criterios de parada
max_depth = 6
num_pct = 6
max_features = None
min_gain=0.001
```

```python
# Separación predictoras y target
X1 = bikes[['season', 'hour']].reset_index(drop=True)
y1 = bikes['total'].reset_index(drop=True).astype(int)

X,val_X,y,val_y = train_test_split(X1,y1,test_size=0.3)
```

```python
# Impresión variable a usar (hour)
j = 1
print(X.columns[j])
```

```python
# División de la variable hour en num_ctp puntos (parámetro definido anteriormente) para obtener posibles puntos de corte
splits = np.percentile(X.iloc[:, j], np.arange(0, 100, 100.0 / num_pct).tolist())
splits = np.unique(splits)
splits
```

```python
# División de las observaciones usando el punto de corte en la posición 5 de la lista de splits
k=5
filter_l = X.iloc[:, j] < splits[k]

# División de la variable de respuesta de acuerdo a si la observación cumple o no con la regla binaria
# y_l: la observación tiene un valor menor al punto de corte seleccionado
# y_r: la observación tiene un valor mayor o igual al punto de corte seleccionado
y_l = y.loc[filter_l]
y_r = y.loc[~filter_l]
```

```python
# Definición de la función que calcula el gini index
def gini(y):
    if y.shape[0] == 0:
        return 0
    else:
        return 1 - (y.mean()**2 + (1 - y.mean())**2)
```

```python
# Gini index de las observaciones que tienen un valor menor al punto de corte seleccionado
gini_l = gini(y_l)
gini_l
```

```python
# Gini index de las observaciones que tienen un valor mayor o igual al punto de corte seleccionado
gini_r = gini(y_r)
gini_r
```

```python
# Definición de la función gini_imputiry para calular la ganancia de una variable predictora j dado el punto de corte k
def gini_impurity(X_col, y, split):
    
    filter_l = X_col < split
    y_l = y.loc[filter_l]
    y_r = y.loc[~filter_l]
    
    n_l = y_l.shape[0]
    n_r = y_r.shape[0]
    
    gini_y = gini(y)
    gini_l = gini(y_l)
    gini_r = gini(y_r)
    
    gini_impurity_ = gini_y - (n_l / (n_l + n_r) * gini_l + n_r / (n_l + n_r) * gini_r)
    
    return gini_impurity_
```

```python
# Ganancia de la variable 'hour' en el punto de corte selecionado
gini_impurity(X.iloc[:, j], y, splits[k])
```

```python
# Todas las particisones
# Definición de la función best_split para calcular cuál es la mejor variable y punto de cortepara hacer la bifurcación del árbol
def best_split(X, y, num_pct=10):
    
    features = range(X.shape[1])
    
    best_split = [0, 0, 0]  # j, split, gain
    
    # Para todas las varibles 
    for j in features:
        
        splits = np.percentile(X.iloc[:, j], np.arange(0, 100, 100.0 / (num_pct+1)).tolist())
        splits = np.unique(splits)[1:]
        
        # Para cada partición
        for split in splits:
            gain = gini_impurity(X.iloc[:, j], y, split)
                        
            if gain > best_split[2]:
                best_split = [j, split, gain]
    
    return best_split
```

```python
# Obtención de la variable 'j', su punto de corte 'split' y su ganancia 'gain'
j, split, gain = best_split(X, y, 5)
j, split, gain
```

```python
# División de las observaciones usando la mejor variable 'j' y su punto de corte 'split'
filter_l = X.iloc[:, j] < split

y_l = y.loc[filter_l]
y_r = y.loc[~filter_l]
```

```python
y.shape[0], y_l.shape[0], y_r.shape[0]
```

```python
y.mean(), y_l.mean(), y_r.mean()
```

```python
# Definición de la función tree_grow para hacer un crecimiento recursivo del árbol
def tree_grow(X, y, level=0, min_gain=0.001, max_depth=None, num_pct=10):
    
    # Si solo es una observación
    if X.shape[0] == 1:
        tree = dict(y_pred=y.iloc[:1].values[0], y_prob=0.5, level=level, split=-1, n_samples=1, gain=0)
        return tree
    
    # Calcular la mejor división
    j, split, gain = best_split(X, y, num_pct)
    
    # Guardar el árbol y estimar la predicción
    y_pred = int(y.mean() >= 0.5) 
    y_prob = (y.sum() + 1.0) / (y.shape[0] + 2.0)  # Corrección Laplace 
    
    tree = dict(y_pred=y_pred, y_prob=y_prob, level=level, split=-1, n_samples=X.shape[0], gain=gain)
    # Revisar el criterio de parada 
    if gain < min_gain:
        return tree
    if max_depth is not None:
        if level >= max_depth:
            return tree   
    
    # Continuar creando la partición
    filter_l = X.iloc[:, j] < split
    X_l, y_l = X.loc[filter_l], y.loc[filter_l]
    X_r, y_r = X.loc[~filter_l], y.loc[~filter_l]
    tree['split'] = [j, split]

    # Siguiente iteración para cada partición
    
    tree['sl'] = tree_grow(X_l, y_l, level + 1, min_gain=min_gain, max_depth=max_depth, num_pct=num_pct)
    tree['sr'] = tree_grow(X_r, y_r, level + 1, min_gain=min_gain, max_depth=max_depth, num_pct=num_pct)
    
    return tree
```

```python
# Aplicación de la función tree_grow
tree_grow(X, y, level=0, min_gain=0.001, max_depth=6, num_pct=10)
```

```python
tree = tree_grow(X, y, level=0, min_gain=0.001, max_depth=6, num_pct=10)
tree
```

```python
# Definición de la función tree_predict para hacer predicciones según las variables 'X' y el árbol 'tree'

def tree_predict(X, tree, proba=False):
    
    predicted = np.ones(X.shape[0])

    # Revisar si es el nodo final
    if tree['split'] == -1:
        if not proba:
            predicted = predicted * tree['y_pred']
        else:
            predicted = predicted * tree['y_prob']
            
    else:
        
        j, split = tree['split']
        filter_l = (X.iloc[:, j] < split)
        X_l = X.loc[filter_l]
        X_r = X.loc[~filter_l]

        if X_l.shape[0] == 0:  # Si el nodo izquierdo está vacio solo continua con el derecho 
            predicted[~filter_l] = tree_predict(X_r, tree['sr'], proba)
        elif X_r.shape[0] == 0:  #  Si el nodo derecho está vacio solo continua con el izquierdo
            predicted[filter_l] = tree_predict(X_l, tree['sl'], proba)
        else:
            predicted[filter_l] = tree_predict(X_l, tree['sl'], proba)
            predicted[~filter_l] = tree_predict(X_r, tree['sr'], proba)

    return predicted
```

```python
# Ejecución de función tree_predict
y_pred = tree_predict(val_X, tree)
```

```python
# Cálculo del error del modelo
np.sqrt(mean_squared_error(val_y, y_pred))
```

### Punto 8 - Votación Ponderada
En la celda 8 elabore un ensamble con la metodología de **Votación ponderada** compuesto por 300 muestras bagged para los mismos tres escenarios del punto 7. Evalúe los modelos utilizando las métricas de Accuracy y F1-Score

```python
# Celda 8
# Obtención de los pesos alpha de cada modelo de acuerdo al error OOB - VP1
errors1 = np.zeros(clf1.n_estimators)
y_pred_all_1 = np.zeros((X_test.shape[0], clf1.n_estimators))

for i in range(clf1.n_estimators):
    oob_sample = ~clf1.estimators_samples_[i]
    y_pred_ = clf1.estimators_[i].predict(X_train.values[oob_sample])
    errors1[i] = metrics.accuracy_score(y_pred_, y_train.values[oob_sample])
    y_pred_all_1[:, i] = clf1.estimators_[i].predict(X_test)
    
alpha1 = (1 - errors1) / (1 - errors1).sum()
y_pred1 = (np.sum(y_pred_all_1 * alpha1, axis=1) >= 0.5).astype(np.int)

# Desempeño al hacer votación ponderada
Accuracy_VP1 = metrics.accuracy_score(y_pred1, y_test)
F1_Score_VP1 = metrics.f1_score(y_pred1, y_test)

print('Accuracy: ' + str(Accuracy_VP1))
print('F1-Score: ' + str(F1_Score_VP1))
```

```python
# Obtención de los pesos alpha de cada modelo de acuerdo al error OOB -VP3
errors2 = np.zeros(clf2.n_estimators)
y_pred_all_2 = np.zeros((X_test.shape[0], clf2.n_estimators))

for i in range(clf2.n_estimators):
    oob_sample = ~clf2.estimators_samples_[i]
    y_pred_ = clf2.estimators_[i].predict(X_train.values[oob_sample])
    errors2[i] = metrics.accuracy_score(y_pred_, y_train.values[oob_sample])
    y_pred_all_2[:, i] = clf2.estimators_[i].predict(X_test)
    
alpha2 = (1 - errors2) / (1 - errors2).sum()
y_pred2 = (np.sum(y_pred_all_2 * alpha2, axis=1) >= 0.5).astype(np.int)

# Desempeño al hacer votación ponderada
Accuracy_VP2 = metrics.accuracy_score(y_pred2, y_test)
F1_Score_VP2 = metrics.f1_score(y_pred2, y_test)

print('Accuracy: ' + str(Accuracy_VP2))
print('F1-Score: ' + str(F1_Score_VP2))
```
